```python
# -Write a function cube() that raises an argument to the third power and returns that argument
# -> include a docstring in the function and include a call to cube()
def cube(number):
    '''Function takes in a float or integer and raises that value to the third power
    and then returns that value.'''
    number **= 3
    return number
# create a list of 5 numbers..
numbers = [1, 2, 3, 4, 5]
# output the numbers fir user to see
print(f'List of numbers:\n{numbers}')
# create a list of the numbers cubed by using list comprehension and calling our function
numbersCubed = [cube(number) for number in numbers]
print(f'List of the numbers cubed:\n{numbersCubed}')
```
