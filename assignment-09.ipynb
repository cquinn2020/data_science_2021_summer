# Requirement 4: Explain how NumPy arrays are like Python lists and how they are different.

# Answer:  Both NumPy arrays and Python lists are mutable objects. Both have built in objects that can be used to manipulate and work with the data. One major difference is that Python lists can contain different data types, while in a NumPy array all the elements have to have the same data type. You can still iterate through both objects and perform computations; however, computing will be faster and more efficient using the NumPy library because of the fixed dtype.

# Requirement 6: Explain how integer storage is performed in Python compared to how it is performed in a language like C.

# Answer: The current Python implementation is constructed in C and every Python object is really a C structure. A C integer is basically a label for a position in memory whose bytes encode an integer value. A Python integer is a pointer to a position in memory containing all the Python object information such as the bytes containing the integer value.

# Requirement 10: Create NumPy array of 10 floats named one_d_array. Demonstrate the following operations on the array: implicit and explicit casting

```python
# create the one_d_array
one_d_array = np.random.random_sample(size=10)
print(f'Original one_d_array:\n{one_d_array}')
# perform explicit operation -> python converts dtype into another dtype automatically
one_d_array = [int(num) for num in one_d_array]
print(f'one_d_array after explicit conversion to type int: {one_d_array}')
# perform implicit casting
one_d_array[0] = 9.0
print(f'one_d_array after implicit conversion of first element to type float: {one_d_array}')
```

# Requirement 11: Use list comprehension to create a two-dimensional NumPy array named two_d_array with 5 rows and 3 col with the following content

```python
import numpy as np


def merge_sum(a, b):
    # assume a and b are same length
    assert len(a) == len(b)
    return [x + b[i] for i, x in enumerate(a)]


two_d_array = np.array([merge_sum([row, row, row], [0, 1, 2]) for row in range(1, 6)])

print(two_d_array)
```

# Requirement 23:

```python
xValues = np.arange(-1e6, 1e6, 1)
def computeFunc(xvals):
    y = []
    for x in xvals:
        y.append(x ** 2 + 56)
    return np.array(y)
yVals = computeFunc(xValues)
print(f'The sum of the y values is: {np.sum(yVals)}')
print(f'The minimum of the y values is: {np.min(yVals)}')
print(f'The maximum of the y values is: {np.max(yVals)}')
print("Using %timeit magic function to time np.sum, np.min, np.max functions:")
%timeit np.sum(yVals)
%timeit np.min(yVals)
%timeit np.max(yVals)
print("Using %timeit magic function to time built in sum, min, max functions:")
%timeit sum(yVals)
%timeit min(yVals)
%timeit max(yVals)
print('Working with the NumPy array appears to be much more efficient and faster!')
```
