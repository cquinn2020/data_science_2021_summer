```python
import numpy as np
import pandas as pd

class display(object):
    """Display HTML representation of multiple objects"""
    template = """<div style="float: left; padding: 10px;">
    <p style='font-family:"Courier New", Courier, monospace'>{0}</p>{1}
    </div>"""
    def __init__(self, *args):
        self.args = args
        
    def _repr_html_(self):
        return '\n'.join(self.template.format(a, eval(a)._repr_html_())
                         for a in self.args)
    
    def __repr__(self):
        return '\n\n'.join(a + '\n' + repr(eval(a))
                           for a in self.args)
```

```python
# 5. Use the following seeds for RandomState() to demonstrate that the function
# returns the same random numbers for constant seeds:
# • RandomState(42)
# • RandomState(99)
# • RandomState(42)
# • RandomState(99)
def show_same_random(seed_num):   
    
    if seed_num == 42:
        print()
        rng = np.random.RandomState(42)
        ser1 = pd.Series(rng.rand(5))
        ser2 = pd.Series(rng.rand(5))
        print(f'Series 1:\n{ser1}')
        print(f'Series 2:\n{ser2}')
    
    elif seed_num == 99:
        print()
        rng = np.random.RandomState(99)
        ser3 = pd.Series(rng.rand(5))
        ser4 = pd.Series(rng.rand(5))
        print(f'Series 3:\n{ser3}')
        print(f'Series 4:\n{ser4}')
    
    else:
        print('Not a valid seed number.')

        
print('Output for running RandomState() with seed 42 twice:')    
show_same_random(42)
show_same_random(42)
print('\nOutput for running RandomState() with seed 99 twice:') 
show_same_random(99)
show_same_random(99)
```

# Requirement 21:

```python
# 21. Demonstrate the use of five Pandas str methods on a Pandas series named
# states created from a list of seven names of U.S. states.

states = pd.Series(['Kentucky', 'Michigan', 'Texas', 'Alabama', 'Florida', 'Tennessee', 'North Carolina'])

def use_str_methods(ser):
    print(ser)
    print(ser.str.upper())
    print(ser.str.lower())
    print(ser.str.split())
    print(ser.str.ljust(100, fillchar='*'))
    print(ser.str.rjust(100, fillchar='*'))
    print(ser.str.find('Michigan'))
    
    
use_str_methods(states)
```

```python
# 28. Use the timeit function to compare processes summing four Pandas
# dataframe objects using:
# • the typical approach (simple ‘+’ operations)
# • the pd.eval() approach
nrows, ncols = 100000, 100
rng = np.random.RandomState(42)
df1, df2, df3, df4 = (pd.DataFrame(rng.rand(nrows, ncols))
                      for i in range(4))
%timeit df1 + df2 + df3 + df4
%timeit pd.eval('df1 + df2 + df3 + df4')
print('The pd.eval() is about twice as fast like the author mentioned!')
```
